### No 5a. Drop columns which has missing value that cannot be filled with any logical value (like name, host_id, etc)

```python
import numpy as np
```

```python
data.isna().sum()
```

```python
missing_values = ["?", "--", "|||"]
data2 = pd.read_csv("datasets/airbnb.csv", na_values = missing_values)
```

```python
data2
```

```python
data2.isna().sum()
```

```python
data2.dropna(axis=1, how='all', inplace=True)
data2
```

```python
data2.dropna(axis=1, how='any')
```

### No 5b. Fill the missing value in last_review with the most common date

```python
data2
```

```python
data2['last_review'] = data2['last_review'].fillna(data2['last_review'].mode()[0])
data2
```

### 5c. Fill the missing value in reviews_per_month with the mean

```python
data2.iloc[:, 13] = data2.iloc[:, 13].fillna(data2.iloc[:, 13].mean())
data2
```

### No 6. Make category based on the price: Cheap, Medium, and Expensive

```python
data2["price_cat"] = np.where(data2["price"] > 6000, "Expensive", np.where(data2["price"] < 3000, "Cheap", "Medium"))
data2
```

### No 8. Visualize the price to compare the price of available rooms

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data2[["price", "availability_365"]]
```

```python
availability_x = data2['availability_365']
price_y = data2['price']
```

```python
plt.scatter(availability_x, price_y)

plt.title('Price vs Availability')
plt.xlabel('Availability')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
plt.scatter(availability_x, price_y)

plt.xscale('log')
plt.yscale('log')

plt.title('Price vs Availability')
plt.xlabel('Availability')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```
